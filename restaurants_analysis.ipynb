```python
# prompt: relate average meal price ,marketing budget and revenue and marketing budget percentage from revenue

import matplotlib.pyplot as plt
# Calculate Marketing Budget Percentage from Revenue
data['Marketing Budget Percentage'] = (data['Marketing Budget'] / data['Revenue']) * 100

# Select relevant columns
subset = data[['Average Meal Price', 'Marketing Budget', 'Revenue', 'Marketing Budget Percentage']]

# Calculate correlation matrix
correlation_matrix = subset.corr()

# Display correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Scatter plots for pairwise relationships
sns.pairplot(subset)
plt.show()

print("Observations from the correlation matrix and scatter plots:")
print("- Positive correlation between Average Meal Price, Marketing Budget, and Revenue.")
print("- Higher average meal prices tend to be associated with higher marketing budgets and revenue.")
print("- Marketing Budget Percentage shows a varied relationship, suggesting that the effectiveness of marketing spend might differ across restaurants.")
```

```python
# prompt: predict future revenue with the findings if we increase capacity  of each restaurant by 20 percent what is the revenue increase percentage

# Calculate the increased seating capacity
data['Increased Seating Capacity'] = data['Seating Capacity'] * 1.2

# Assuming a linear relationship between seating capacity and revenue,
# estimate the revenue increase proportionally
data['Predicted Revenue Increase'] = (data['Increased Seating Capacity'] / data['Seating Capacity']) * data['Revenue']

# Calculate the percentage increase in revenue
data['Revenue Increase Percentage'] = ((data['Predicted Revenue Increase'] - data['Revenue']) / data['Revenue']) * 100

# Display the results
print(data[['Seating Capacity', 'Increased Seating Capacity', 'Revenue', 'Predicted Revenue Increase', 'Revenue Increase Percentage']])

# Calculate and print the average revenue increase percentage
average_revenue_increase = data['Revenue Increase Percentage'].mean()
print("\nAverage Predicted Revenue Increase Percentage: {:.2f}%".format(average_revenue_increase))
```

```python
# prompt: make graph

import matplotlib.pyplot as plt
# Assuming you want to visualize the 'Revenue Increase Percentage' for each restaurant
plt.figure(figsize=(10, 6))
plt.bar(data.index, data['Revenue Increase Percentage'])
plt.xlabel('Restaurant Index')
plt.ylabel('Revenue Increase Percentage')
plt.title('Predicted Revenue Increase Percentage by Increasing Seating Capacity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
